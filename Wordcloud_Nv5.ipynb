#InstalaÃ§Ã£o e ImportaÃ§Ã£o de Bibliotecas

```python
!pip install Faker -q 
!pip install wordcloud -q 
!pip install num2words -q

from matplotlib import pyplot as plt
from faker import Faker
from random import randint
from collections import Counter
from num2words import num2words
from wordcloud import WordCloud
```

#GeraÃ§Ã£o de GravaÃ§Ã£o de dados em txt.



```python
n = int(input('Quantos dados serÃ£o gerados? '))

# Gerar nomes e nÃºmeros aleatÃ³rios dentro de um arquivo txt
fake = Faker()
with open('dados.txt', 'w') as arquivo:
    for x in range(0, n):
        arquivo.write(f'{fake.name()},{(randint(1, 10))},')
```

#RecuperaÃ§Ã£o e ManipulaÃ§Ã£o dos dados

```python
ler = []
with open('dados.txt', 'r') as arquivo:
    ler = arquivo.read().split(',')
ler.pop()  

nomes = []
notas = []
for x in range(0, (n*2), 2):
    nomes.append(ler[x])
    notas.append(ler[x + 1])

num = []
for x in notas:
    num.append(num2words(x, lang='pt-br').title())

palavras = " ".join(p for p in num)
wordcloud = WordCloud(background_color="black").generate(palavras)
```

#ImpressÃ£o dos dados


```python

#Histograma
notas.sort()
plt.hist(notas)
plt.title('Histograma das pontuaÃ§Ãµes', fontsize=15)
plt.xlabel('PontuaÃ§Ãµes')
plt.ylabel('Probabilidade')
plt.grid(f'{Counter(notas)}')
#Wordcloud
plt.subplots()
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(0)
plt.show()
```
